# NumCompute Stream Demo

This notebook demonstrates the streaming machine learning framework built using only Python, NumPy, and matplotlib.

The demo performs the following steps:

1. Loads a CSV dataset using the custom `io.py` module.
2. Splits the dataset into chunks to simulate streaming data.
3. Trains a Decision Tree pipeline incrementally.
4. Trains a Random Forest pipeline incrementally.
5. Logs accuracy, error, memory usage, and cumulative accuracy.
6. Visualises model performance over time.


```python
import os
import sys

# Add project root to Python path
sys.path.insert(0, os.path.abspath('..'))

import numpy as np

from numcompute_stream.io import load_csv, make_chunks
from numcompute_stream.preprocessing import SimpleImputer, StandardScaler
from numcompute_stream.tree import DecisionTreeClassifier
from numcompute_stream.ensemble import RandomForestClassifier
from numcompute_stream.pipeline import Pipeline
from numcompute_stream.stream import StreamTrainer
from numcompute_stream.visualise import (
    plot_metric_over_time,
    compare_models,
    plot_predictions_vs_ground_truth,
    plot_error_over_time,
    plot_memory_usage,
)
```

## 1. Load Dataset

The dataset is loaded using the custom CSV loader from `io.py`. No pandas or external data-processing libraries are used.

```python
dataset_path = 'sample_dataset.csv'

X, y = load_csv(dataset_path, target_column=-1, skip_header=True)

print('X shape:', X.shape)
print('y shape:', y.shape)
print('First 5 rows of X:')
print(X[:5])
print('First 5 labels:')
print(y[:5])
```

## 2. Create Streaming Chunks

The dataset is split into chunks to simulate a real-time streaming learning scenario.

```python
chunks = make_chunks(X, y, chunk_size=8)
print('Number of chunks:', len(chunks))

for i, (X_chunk, y_chunk) in enumerate(chunks):
    print(f'Chunk {i}: X={X_chunk.shape}, y={y_chunk.shape}')
```

## 3. Build Streaming Pipelines

Two pipelines are created:

- A single Decision Tree pipeline
- A Random Forest ensemble pipeline

Both use a `SimpleImputer` and `StandardScaler` before the model.

```python
tree_pipeline = Pipeline([
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler()),
    ('model', DecisionTreeClassifier(
        max_depth=4,
        min_samples_split=2,
        criterion='gini'
    )),
])

forest_pipeline = Pipeline([
    ('imputer', SimpleImputer()),
    ('scaler', StandardScaler()),
    ('model', RandomForestClassifier(
        n_estimators=7,
        max_depth=4,
        min_samples_split=2,
        criterion='gini',
        random_state=42
    )),
])

tree_trainer = StreamTrainer(tree_pipeline, verbose=True)
forest_trainer = StreamTrainer(forest_pipeline, verbose=True)
```

## 4. Train Decision Tree Incrementally

Each chunk is passed to `.fit_chunk()`, which internally calls `.partial_fit()`.

```python
for X_chunk, y_chunk in chunks:
    tree_trainer.fit_chunk(X_chunk, y_chunk)

tree_logs = tree_trainer.get_logs()
tree_logs
```

## 5. Train Random Forest Incrementally

The Random Forest is trained in the same streaming way. It uses multiple decision trees and majority voting.

```python
for X_chunk, y_chunk in chunks:
    forest_trainer.fit_chunk(X_chunk, y_chunk)

forest_logs = forest_trainer.get_logs()
forest_logs
```

## 6. Final Accuracy Results

```python
print('Decision Tree final cumulative accuracy:', tree_logs['cumulative_accuracy'][-1])
print('Random Forest final cumulative accuracy:', forest_logs['cumulative_accuracy'][-1])
```

## 7. Visualise Accuracy Over Time

```python
plot_metric_over_time(
    tree_logs['accuracy'],
    title='Decision Tree Accuracy Over Streaming Chunks',
    ylabel='Accuracy'
)
```

```python
plot_metric_over_time(
    forest_logs['accuracy'],
    title='Random Forest Accuracy Over Streaming Chunks',
    ylabel='Accuracy'
)
```

## 8. Compare Models

```python
compare_models(
    tree_logs['accuracy'],
    forest_logs['accuracy'],
    labels=('Decision Tree', 'Random Forest'),
    title='Streaming Accuracy Comparison',
    ylabel='Accuracy'
)
```

## 9. Plot Error and Memory Usage

```python
plot_error_over_time(
    forest_logs['error'],
    title='Random Forest Error Over Streaming Chunks'
)
```

```python
plot_memory_usage(
    forest_logs['memory_bytes'],
    title='Random Forest Memory Footprint Per Chunk'
)
```

## 10. Predictions vs Ground Truth

The final chunk is used to compare predicted labels against actual labels.

```python
latest_X, latest_y = chunks[-1]
latest_pred = forest_trainer.predict(latest_X)

print('Ground truth:', latest_y.astype(int))
print('Predictions :', latest_pred.astype(int))

plot_predictions_vs_ground_truth(
    latest_y,
    latest_pred,
    title='Latest Chunk: Predictions vs Ground Truth'
)
```

## Summary

This notebook demonstrates that the NumCompute Stream framework supports chunk-wise data processing, incremental model updates, streaming metrics, model comparison, and visualisation using only NumPy and matplotlib.
